```python
n = 1000
count = 0
for i in range(n):
    for j in range(n):
        for k in range(n):
            if i < j and j < k:
                count += 1
print(count)
```

Above, we are counting all triples of different integers _i_, _j_, _k_ such that `1 <= i < j < k <= 1000`.
It is the same as the number of ways of selecting three different integers out of 1000 since one can always arrange them in decreasing order. Thus, the answer is `1000 choose 3` which is equal to _166167000_.

```python
# What is the number of non-negative integers with at most four digits at least one of which is equal to 7?

from itertools import product

count = 0
for d in product(range(10), repeat=4):
    if 7 in d:
        count += 1
print(count)
```

```python
# What is the number of non-negative integers with at most four digits whose digits are increasing?

from itertools import product

count = 0
for d in product(range(10), repeat=4):
    if d[0] < d[1] and d[1] < d[2] and d[2] < d[3]:
        count += 1
print(count)
```

```python
# What is the number of bit-strings (that is, strings consisting of 0's and 1's) of length 6 where 
# the number of 0's is equal to the number of 1's?

from itertools import product

counter = 0
for i in product('01', repeat=6):
    if i.count('0') == i.count('1'):
        counter += 1
print(counter)
```

```python
# What is the number of non-negative integers with at most six digits 2 of them even and 3 of them odd?

from itertools import product

counter = 0
for number in product(range(10), repeat=6):
    # The sum of the digits have to be odd(odd + odd + odd + even + even + even = odd)
    if sum(number) % 2 == 0:
        continue
    # now check the current tuple's digits whether there are 3 even 3 odd numbers
    odd = 0
    even = 0
    for char in number:
        if char % 2 == 0:
            even += 1
        else: 
            odd += 1
    if odd == even:
        counter += 1
print(counter)
    
```

### REVIEW

* Below we are gonna show four standard types of selections.

* To visualize four of them, we use an example of selecting _k=2_ objects out of a set _{1, 2, 3}_ of size n=3.

* We distinguish ordered and unordered selections. In mathematics, the ordered ones are written in paranthesis(that is, round brackets) whereas the unordered ones are written in braces(that is, curly braces).

* Shortly, we treat _(c, a)_ and _(a, c)_ as different, however we consider _{c, a}_ and _{a, c}_ as being the same.

* We also distinguish selections with and without repetitions: _(c, c)_ contains a repeating element, whereas _(c, a)_ does not.

```python
# Tuples are ordered selections with repetitions. Also known as words. The number of tuples is n^k
from itertools import product

for t in product('abc', repeat=2):
    print(*t, sep='', end=' ')
```

```python
# Permutations are ordered selections without repetitions. The number of permutations n!/(n-k)!
from itertools import permutations

for t in permutations('abc', 2):
    print(*t, sep='', end=' ')
```

```python
# Combinations are unordered selections without repetitions. Also known as sets. n!/k!x(n-k)!
from itertools import combinations

for t in combinations('abc', 2):
    print(*t, sep='', end=' ')
```

### Problems in Combinatorics

```python
# There are 9 assignments and 4 people. How many ways are there to give a single assignment to every person (so that
# different people receive different assignments)?

# The explanation of the solution: Give the first person any of 9 assignments, give the second person any of the 
# remaining eight assignments and so on. Hence, by the product rule:
from itertools import permutations

for i, d in enumerate(permutations(range(9), 4)):
    print(str(i) + ': ', d)
```

```python
# There are 15 identical candies. How many ways are there to distribute them among 7 kids?

# The explanation of the solution:For every of k=15 candies, we select one of n=7 kids that receives this candy.
# Any kid may be selected any number of times (including zero number of times). 
# Thus, we are dealing with k-combinations with repetitions selected out of n objects. This gives the answer:  
from itertools import combinations_with_replacement

counter = 0
for c in combinations_with_replacement('ABCDEFG', 15):
    counter += 1
print(counter)
```

```python
# There are 15 identical candies. How many ways are there to distribute them among 7 kids in such a way that each kid 
# receives at least 1 candy?

# The explanation of the solution: Let's give each kid 1 candy. What is left is 8 candies. The question simplifies to
# How many ways are there to distribute 8 candies to 7 children

from itertools import combinations_with_replacement

counter = 0
for c in combinations_with_replacement('ABCDEFG', 8):
    counter += 1
print(counter)
```

```python
# How many non-negative integer numbers are there below 10000 such that their sum of digits is equal to 9?

# We are gonna use the import method as we want to include every possible sequence of 6 digit numbers
from itertools import product

counter = 0
for d in product(range(10), repeat=4):
    if sum(d) == 9:
        counter += 1
print(counter)
```

```python
# How many four-digt numbers are there such that their digits are non-increasing? Three-digit numbers are also 
# four digit, they just start with 0.

from itertools import product

counter = 0
for d in product(range(10), repeat=4):
    if d[0] >= d[1] and d[1] >= d[2] and d[2] >= d[3]:
        counter += 1
counter
```

* 6 people including A, B and C form a queue in a random order. (All 6! orderings are equiprobable). Consider the 
  event **_A is the first in the queue_**. What is the probability?

```python
# this means that the index of A(if we imagine the queue as a tuple or list) is 0.
from itertools import permutations

perms = list(permutations('ABCDEF'))

wanted_events = 0
for i in perms:
    index_a = i.index('A')
    if index_a == 0:
        wanted_events += 1
all_events = 6*5*4*3*2*1
the_probability = wanted_events / all_events
print(the_probability)
```

- 6 people, including A,B, and C, form a queue in a random order (all 6! orderings are equiprobable). Consider the event **_A precedes B in the queue_**. What is the probability?

```python
# This means that the index number of A must always be lowet than index number of B.
from itertools import permutations

perms = list(permutations('ABCDEF'))

wanted_events = 0
for i in perms:
    index_a = i.index('A')
    index_b = i.index('B')
    if index_b > index_a:
        wanted_events += 1
all_events = 6*5*4*3*2*1
the_probability = wanted_events / all_events
print(the_probability)
```

- 6 people, including A,B, and C, form a queue in a random order (all 6! orderings are equiprobable). Consider the event **_B is between A and C in the queue_**. What is the probability?

```python
from itertools import permutations

perms = list(permutations('ABCDEF'))

unwanted_events = 0
for i in perms:
    index_a = i.index('A')
    index_b = i.index('B')
    index_c = i.index('C')
    # We are gonna eliminate the situations that B is not in the middle of the A and C
    if (index_b > index_a and index_b > index_c) or (index_b < index_a and index_b < index_c): 
        unwanted_events += 1
all_events = 6*5*4*3*2*1
wanted_events = all_events - unwanted_events
the_probability = wanted_events / all_events
print(the_probability)
```

- 6 people, including A,B, and C, form a queue in a random order (all 6! orderings are equiprobable). Consider the event **_A and B are neighbors in the queue_**. What is the probability?

```python
# This means that (the index number of A) - (the index number of B) = 1
from itertools import permutations

perms = list(permutations('ABCDEF'))

wanted_events = 0
for i in perms:
    index_a = i.index('A')
    index_b = i.index('B')
    if abs(index_b - index_a) == 1:
        wanted_events += 1
all_events = 6*5*4*3*2*1
the_probability = wanted_events / all_events
print(the_probability)
```

- A king decides to give a prisoner a chance of being set free. The kings places two boxes, **_15 white_** and **_15   black_** balls on the table. The prisoner can distribute the balls between two boxes in any way with the only 
  constraint that no box is empty. After this, the king picks one of the boxes with probability 1/2 each, and picks   a random ball from that box (each ball is picked with equal probability). If the ball happens to be **white**,       then the prisoner is freed. Distribute the balls between the two boxes to **maximize** prisoner’s chances.

```python
max_probability = 0

for i in range(15, -1, -1):
    for j in range(14, 0, -1):
        box1_white = i
        box1_black = j
        box2_white = 15 - i
        box2_black = 15 - j
        temp_probability = box1_white / (box1_white + box1_black)
        if temp_probability > max_probability:
            max_probability = temp_probability
print(max_probability)
```

- 6 people form a queue, including A and B. What is the probability that B is before A under the condition that A 
  is not the first?

```python
from itertools import permutations

perms = list(permutations('ABCDEF'))

wanted_events = 0
a_not_first = 0
for i in perms:
    index_a = i.index('A')
    index_b = i.index('B')
    if index_a != 0:
        a_not_first += 1
        if index_b < index_a:
            wanted_events += 1
print(wanted_events/a_not_first)
```

```python
from itertools import accumulate
import matplotlib.pyplot as plt
from operator import mul

num_days = 365

factors = [1 - i / num_days for i in range(num_days)]
prob_no_collision = list(accumulate(factors, mul))
prob_collision = [1 - p for p in prob_no_collision]

plt.plot(prob_collision)

for n in (22, 23, 35, 100):
    print(f'The probability for {n} people '
          f'is {prob_collision[n - 1]}')
```

### Unconventional Dice Game

- Alice and Bob play the following game. Each of them has fair but unconventional dice. 
- The numbers on the sides of 
  - **Alice's dice are: 2,2,2,2,3,3**
  - **Bob's dice are: 1,1,1,1,6,6**

- If Alice and Bob play the game many times, who is likely to win more often?

Let us compute the expectations of the corresponding two random variables. Alice gets 2 with probability 2/3 and gets 3 with probability 1/3. Her expectation is 

- 2x2/3 + 3x1/3 = 7/3

Bob gets 1 with probability 2/3 and gets 6 with probability 1/3. His expected value is  

- 1x2/3 + 6x1/3 = 8/3

Therefore, Bob has better expected value. Does it mean that he wins more often?

Let us compute the probabilities for each player to win directly. The result of the game depends only on the outcome of Bob: indeed, if he throws 1, he loses, if he throws 6, he wins. Bob throws 1 with probability 2/3. Thus, Bob loses with probability 2/3. Bob loses more often despite having larger expected value. 

The following code snippet simulates the game 10**5 times and shows that, indeed, Alice wins in about 2/3 of all games.

```python
from random import randint, seed

seed(25)

dice1 = [2, 2, 2, 2, 3, 3]
dice2 = [1, 1, 1, 1, 6, 6]

num_rounds = 10 ** 5
num_dice1_wins = 0

for _ in range(num_rounds):
    dice1_result = dice1[randint(0, 5)]
    dice2_result = dice2[randint(0, 5)]
    
    if dice1_result > dice2_result:
        num_dice1_wins += 1
print(f'Out of {num_rounds} throws, dice1 won {num_dice1_wins} times')
```
